### Preprocessing

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
```

```python
x = df.drop(columns=['target'])
y = df['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)

ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report 
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred_lr = lr.predict(x_test)

conf_matrix = confusion_matrix(y_test, y_pred_lr)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
class_report = classification_report(y_test, y_pred_lr, output_dict=True)

precision_lr = class_report['1']['precision']
recall_lr = class_report['1']['recall']
f1_lr = class_report['1']['f1-score']

print("Confusion Matrix:\n", conf_matrix)
print("Accuracy Score:", accuracy_lr)
print("F1 Score:", f1_lr)
print("Precision:", precision_lr)
print("Recall:", recall_lr)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier , export_graphviz
```

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train,y_train)
y_pred_dt = dt.predict(x_test)

conf_matrix = confusion_matrix(y_test, y_pred_dt)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
class_report = classification_report(y_test, y_pred_dt, output_dict=True)

precision_dt = class_report['1']['precision']
recall_dt = class_report['1']['recall']
f1_dt = class_report['1']['f1-score']

print("Confusion Matrix:\n", conf_matrix)
print("Accuracy Score:", accuracy_dt)
print("F1 Score:", f1_dt)
print("Precision:", precision_dt)
print("Recall:", recall_dt)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=500,random_state=42)
rf.fit(x_train,y_train)
y_pred_rf = rf.predict(x_test)

conf_matrix = confusion_matrix(y_test, y_pred_rf)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
class_report = classification_report(y_test, y_pred_rf, output_dict=True)

precision_rf = class_report['1']['precision']
recall_rf = class_report['1']['recall']
f1_rf = class_report['1']['f1-score']

print("Confusion Matrix:\n", conf_matrix)
print("Accuracy Score:", accuracy_rf)
print("F1 Score:", f1_rf)
print("Precision:", precision_rf)
print("Recall:", recall_rf)
```

### XG Boost

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier()
xgb.fit(x_train,y_train)
y_pred_xgb = xgb.predict(x_test)

conf_matrix = confusion_matrix(y_test, y_pred_xgb)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
class_report = classification_report(y_test, y_pred_xgb, output_dict=True)

precision_xgb = class_report['1']['precision']
recall_xgb = class_report['1']['recall']
f1_xgb = class_report['1']['f1-score']

print("Confusion Matrix:\n", conf_matrix)
print("Accuracy Score:", accuracy_xgb)
print("F1 Score:", f1_xgb)
print("Precision:", precision_xgb)
print("Recall:", recall_xgb)
```
